```python
##Define thresholding
def thr(im):
  m= im.shape[0]
  n= im.shape[1]
  temp=0
  im1=np.float32(im)
  for i in range (0,m):
    for j in range(0,n):
      temp += im1[i,j]
      mean=temp/(m*n)
      bin_img=im1
  for i in range (0,m):
    for j in range(0,n):
      if im1[i,j]>=mean:
         bin_img[i,j]=1
      else:
         bin_img[i,j]=0
         bin_img=bin_img.astype(np.int8)
    return bin_img,mean
```

```python
#Formation Blocks
img1=im[0:250,0:250]
img2=im[0:250,251:500]
img3=im[251:500,0:250]
img4=im[251:500:,251:500]
#Plotting of image
plt.figure(figsize=figsize)
plt.imshow(img1,cmap="gray",vmin=0,vmax=256)
plt.title('Block image 1')
plt.show()

plt.figure(figsize=figsize)
plt.imshow(img2,cmap="gray",vmin=0,vmax=256)
plt.title('Block image 2')
plt.show()

plt.figure(figsize=figsize)
plt.imshow(img3,cmap="gray",vmin=0,vmax=255)
plt.title('Block image 3')
plt.show()
plt.figure(figsize=figsize)
plt.imshow(img4,cmap="gray",vmin=0,vmax=255)
plt.title('Block image 4')
plt.show()
```

```python
thr_img1,mm1=thr(img1)
plt.figure(figsize=figsize)
plt.imshow(thr_img1,cmap="gray",vmin=0,vmax=1)
plt.title('Block image 1')
plt.show()
mm1
```

```python
thr_img2,mm2=thr(img2)
plt.figure(figsize=figsize)
plt.imshow(thr_img2,cmap="gray",vmin=0,vmax=1)
plt.title('Block image 2')
plt.show()
mm2
```

```python
thr_img3,mm3=thr(img3)
plt.figure(figsize=figsize)
plt.imshow(thr_img3,cmap="gray",vmin=0,vmax=1)
plt.title('Block image 3')
plt.show()
mm3
```

```python
thr_img4,mm4=thr(img4)
plt.figure(figsize=figsize)
plt.imshow(thr_img4,cmap="gray",vmin=0,vmax=1)
plt.title('Block image 4')
plt.show()
mm1
```

```python
####Local Thresholding#####
####Concatenate the images###
##Horizontal Concated
A=cv2.hconcat([thr_img1,thr_img2])
B=cv2.hconcat([thr_img3,thr_img4])
##vertical Concated
C=cv2.vconcat([A,B])
plt.figure(figsize=figsize)
plt.imshow(C,cmap="gray",vmin=0,vmax=1)
plt.title('Local Thresohlding')
plt.show()
```
