```python
#####Dilation and erosion operation on binary image and grey scale image####
import cv2
import numpy as np
import random
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
from io import BytesIO
from PIL import Image
from google.colab import files
from skimage import color
from skimage import io
#uploaded = files.upload()
from PIL import ImageDraw
from PIL import ImageFont
from skimage.util import random_noise
figsize=(10,10)
```

```python
######Creating Binary Image#####
img=np.zeros([256,256])
img[20:60,20]=1;
img[20:60,22]=1;
img[60:100,60]=1;
img[60:100,62]=1;
img[100:150,100]=1;
img[100:150,102]=1;
img[150:200,150]=1;
img[150:200,152]=1;
img[200:250,200]=1;
img[200:250,202]=1;
img[225:250,250]=1;
img[225:250,255]=1;
plt.figure(figsize=figsize)
plt.imshow(img,cmap="gray",vmin=0, vmax=1)
plt.title("Original Binary image")
plt.show()
```

```python
###Dilation image###
m=img.shape[0]
n=img.shape[1]
mask=np.array([[1,1,1],[1,1,1],[1,1,1]])
img_new=np.zeros([m,n])
for i in range (1,m-1):
  for j in range(1,n-1):
    temp=[img[i-1, j-1]*mask[0, 0],img[i-1, j]*mask[0,1],img[i-1,j+1]*mask[0,2]
          ,img[i,j-1]*mask[1,0],img[i,j]*mask[1,1],img[i,j+1]*mask[1,2],
          img[i+1,j-1]*mask[2,0],img[i+1,j]*mask[2,1],img[i+1,j+1]*mask[2,2]]
    img_new[i,j]=max(temp)
plt.figure(figsize=figsize)
plt.imshow(img_new,cmap='gray',vmin=0,vmax=1)
plt.title('DILATED Image')
plt.show()
```

```python
###Erosion Image#####
m=img.shape[0]
n=img.shape[1]
mask=np.array([[1,1,1],[1,1,1],[1,1,1]])
img_new=np.zeros([m,n])
for i in range (1,m-1):
  for j in range(1,n-1):
    temp=[img[i-1,j-1]*mask[0,0],img[i-1,j]*mask[0,1],img[i-1,j+1]*mask[0,2],
          img[i,j-1]*mask[1,0],img[i,j]*mask[1,1],img[i,j+1]*mask[1,2],
          img[i+1,j-1]*mask[2,0],img[i+1,j]*mask[2,1],img[i+1,j+1]*mask[2,2]] 
    img_new[i,j]=min(temp)
plt.figure(figsize=figsize)
plt.imshow(img_new,cmap="gray",vmin=0, vmax=1)
plt.title("ERODED Image")
plt.show()
```

```python
im= cv2.imread('butterfly.bmp')
img2=cv2.cvtColor(im,cv2.COLOR_BGR2GRAY)
plt.figure(figsize=figsize)
plt.imshow(img2,cmap="gray",vmin=0,vmax=256)
plt.title(" Original Grayscale Image")
plt.show()
```
