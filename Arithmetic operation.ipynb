```python
##Arithmetic Operations on gray scale image###
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
from io import BytesIO
from PIL import Image
from google.colab import files
from skimage import color
from skimage import io
figsize=(10,10)
#upload=files.upload()
```

```python
im1=cv2.imread('cameraman.bmp')
plt.figure(figsize=figsize)
plt.imshow(im1)
plt.title('Cameraman')
plt.show()
im2=cv2.imread('rose.bmp')
plt.imshow(im2)
plt.title('Rose')
plt.show()
```

```python
im3=cv2.resize(im2,(256,256))
im4=cv2.cvtColor(im3,cv2.COLOR_BGR2GRAY)
#cv2_imshow(im3)
#cv2_imshow(im4)
#plt.figure(figsize=figsize)
plt.imshow(im3)
plt.title('Rose Resized Image')
plt.show()
#plt.figure(figsize=figsize)
plt.imshow(im4,cmap='gray')
plt.title('BGR2GRAY Image')
plt.show()
```

```python
R=im3[:,:,0]
G=im3[:,:,1]
B=im3[:,:,2]
#################################
red=im3.copy()
gre=im3.copy()
blue=im3.copy()
###################################
blue[:,:,(1,2)]=0
gre[:,:,(0,2)]=0
red[:,:,(0,1)]=0
planes1=np.concatenate((R,G,B),axis=1)
planes2=np.concatenate((blue,gre,red),axis=1)
plt.figure(figsize=figsize)
plt.imshow(planes1,cmap='gray')
plt.title('R,G,B image')
plt.show()
plt.figure(figsize=figsize)
plt.imshow(planes2,cmap='gray')
plt.title('blue, gre, red image')
plt.show()
```

```python
im5=0.289*B+0.587*G+0.11*R
#cv2_imshow(im5)
#plt.figure(figsize=figsize)
plt.imshow(im5,cmap='gray')
plt.title('Gray image')
plt.show()
```

```python
im6=cv2.add(im5,100)
im7=cv2.subtract(im5,100)
planes3=np.concatenate((im5,im6,im7),axis=1)
#cv2_imshow(planes3)
plt.figure(figsize=figsize)
plt.imshow(planes3,cmap='gray')
plt.title('BGR,addition,substraction image')
plt.show()
```

```python
im8=cv2.multiply(im5,1.5)
im9=cv2.divide(im5,4)
planes4=np.concatenate((im5,im8,im9),axis=1)
#cv2_imshow(planes3)
plt.figure(figsize=figsize)
plt.imshow(planes4,cmap='gray')
plt.title('BGR,Multiply,Division image')
plt.show()
```

```python
im10=cv2.rotate(im6,cv2.cv2.ROTATE_90_CLOCKWISE)
im11=cv2.add(im5,im10)
planes5=np.concatenate((im5,im10,im11),axis=1)
#cv2_imshow(planes4)
plt.figure(figsize=figsize)
plt.imshow(planes5,cmap='gray')
plt.title('rotate image,add image')
plt.show()
```
