```python
####Opening & Closing Morphological operation on binary image###
import cv2
import numpy as np
import random
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
from io import BytesIO
from PIL import Image
from google.colab import files
from skimage import color
from skimage import io
#uploaded = files.upload()
from PIL import ImageDraw
from PIL import ImageFont
from skimage.util import random_noise
figsize=(10,10)
```

```python
####################################OPENING#####################################
######Creating Binary Image######
img=np.zeros([100,100])
img[10:30,10:30]=1;
img[30,30]=1;
img[31:80,31:80]=1;
img[80,80]=1;
img[81:100,81:100]=1;
plt.figure(figsize=figsize)
plt.imshow(img,cmap="gray",vmin=0, vmax=1)
plt.title("Original Binary Image")
plt.show()
```

```python
######Erosion Operation######
m=img.shape[0]
n=img.shape[1]
mask=np.array([[1,1,1],[1,1,1],[1,1,1]])
img_new=np.zeros([m,n])
for i in range (1,m-1):
  for j in range(1,n-1):
    temp=[img[i-1,j-1]*mask[0,0],img[i-1,j]*mask[0,1],img[i-1,j+1]*mask[0,2],
          img[i,j-1]*mask[1,0],img[i,j]*mask[1,1],img[i,j+1]*mask[1,2],
          img[i+1,j-1]*mask[2,0],img[i+1,j]*mask[2,1],img[i+1,j+1]*mask[2,2]]
    img_new[i,j]=min(temp)
    
plt.figure(figsize=figsize)
plt.imshow(img_new,cmap='gray',vmin=0,vmax=1)
plt.title('ERODED Image')
plt.show()
```

```python
######Dilation Operation######
m=img_new.shape[0]
n=img_new.shape[1]
mask=np.array([[1,1,1],[1,1,1],[1,1,1]])
img_new1=np.zeros([m,n])
for i in range (1,m-1):
  for j in range(1,n-1):
    temp=[img_new[i-1,j-1]*mask[0,0],img_new[i-1,j]*mask[0,1],img_new[i-1,j+1]*
          mask[0,2],img_new[i,j-1]*mask[1,0],img_new[i,j]*mask[1,1],
          img_new[i,j+1]*mask[1,2],img_new[i+1,j-1]*mask[2,0],img_new[i+1,j]*
          mask[2,1],img_new[i+1,j+1]*mask[2,2]]
    img_new1[i,j]=max(temp)
plt.figure(figsize=figsize)
plt.imshow(img_new1,cmap='gray',vmin=0,vmax=1)
plt.title('Opening Image')
plt.show()
```

```python
######################CLOSING##########################################
###Generate Binary image
img2=np.zeros([100,100])
img2[10:30,10:30]=1;
img2[29:80,31:80]=1;
img2[79:100,81:100]=1;
plt.figure(figsize=figsize)
plt.imshow(img2,cmap="gray",vmin=0, vmax=1)
plt.title("Original Binary Image")
plt.show()
```

```python
#####Dilation Operation#####
m=img2.shape[0]
n=img2.shape[1]
mask=np.array([[1,1,1],[1,1,1],[1,1,1]])
img_new2=np.zeros([m,n])
for i in range (1,m-1):
  for j in range(1,n-1):
    temp=[img2[i-1,j-1]*mask[0,0],img2[i-1,j]*mask[0,1],img2[i-1,j+1]*
          mask[0,2],img2[i,j-1]*mask[1,0],img2[i,j]*mask[1,1],img2[i,j+1]
          *mask[1,2],img2[i+1,j-1]*mask[2,0],img2[i+1,j]*mask[2,1],img2[i+1,j+1]
          *mask[2,2]]
    img_new2[i,j]=max(temp)
plt.figure(figsize=figsize)
plt.imshow(img_new2,cmap='gray',vmin=0,vmax=1)
plt.title('DILATED')
plt.show()
```

```python
####Erosion operation#####
m=img_new2.shape[0]
n=img_new2.shape[1]
mask=np.array([[1,1,1],[1,1,1],[1,1,1]])
img_new3=np.zeros([m,n])
for i in range (1,m-1):
  for j in range(1,n-1):
    temp=[img_new2[i-1,j-1]*mask[0,0],img_new2[i-1,j]*mask[0,1],
          img_new2[i-1,j+1]*mask[0,2],img_new2[i,j-1]*mask[1,0],img_new2[i,j]
          *mask[1,1],img_new2[i,j+1]*mask[1,2],img_new2[i+1,j-1]*mask[2,0],
          img_new2[i+1,j]*mask[2,1],img_new2[i+1,j+1]*mask[2,2]]
    img_new3[i,j]=min(temp)
    
plt.figure(figsize=figsize)
plt.imshow(img_new3,cmap='gray',vmin=0,vmax=1)
plt.title('Closing Image')
plt.show()
```
